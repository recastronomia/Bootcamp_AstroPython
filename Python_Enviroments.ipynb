En Python existen muchas librerías como `numpy`, `pandas`, `scipy` `...`. Todas ellas se encuentran en constante evolución. Además puede haber interdependecia entre algunas de ellas, es decir, que unas dependan de otras.

Los entornos son espacios aislados donde se pueden gestionar las dependencias necesarias para llevar a cabo nuestro proyectos. Un símil a burbujas cerredas. Esto nos permité dos cosas.

1. Aislamiento de paquetería
2. Portabilidad

En lo que queda de esta sección se usará el modulo `venv` para crear entornos virtuales que ya viene disponible de base en Python desde la versión > `Python 3.3`.

En caso tener una versión anterior, se recomienda ver la documentación de [`virtualvenv`](https://virtualenv.pypa.io/en/latest/).

Miremos algunas versiones por default que tenemos instalada

````python
import numpy
import matplotlib
print('numpy version:', numpy.__version__)
print('matplotlib version:', matplotlib.__version__)
````

Puedo ver la lista de todos los paquetes que tengo instalados usando el siguiente comando.

````python
!pip list
````

Procedemos a crear nuestro primer entorno virtual usando el siguiente comando

```bash
python<version> -m venv <nombre_entorno>
# En mi caso
python -m venv venv_test
```

para **activar** este nuevo ambiente en macOS/linux es:

```bash
source venv_test/bin/activate
```

Ahora veremos que paquetes estan instalados por default en este nuevo ambiente

Para desactivar el ambiente haremos lo siguiente

```bash
deactivate
```

Para crear un archivo de requirimientos

```bash
pip freeze > requirements.txt
```

Para instalar los requirimientos

```bash
pip install ./requirements.txt
```

Ahora si estamos ya dentro del entorno de `conda`
